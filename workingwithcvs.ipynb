```python
import pandas as pd
```

```python
data = {
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'sales': [120, 90, 150, 130, 170, 100]
}
df=pd.DataFrame(data)
df.to_csv('sales.csv',index=False)
print("salese.csv Created")
```

```python
mean_sales=df['sales'].mean()
print("Average Sales:",mean_sales)
```

```python
high_sales=df[df['sales']>mean_sales]
print("Sales higher than average")
print(high_sales)
```

```python
high_sales.to_csv('high_sales.csv',index=False)
print("high_Sales Created")
```
